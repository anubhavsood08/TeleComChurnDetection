```python
import pandas as pd
from sklearn import metrics 
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
# from sklearn.naive_bayes import GaussianNB,MultinomialNB
from sklearn.linear_model import LogisticRegression
from imblearn.combine import SMOTEENN
from sklearn.metrics import accuracy_score,roc_auc_score
```

```python
df= pd.read_csv('df_churn.csv')
df.head(5)
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
df.head(5)
```

```python
# creating x and y variable
x = df.drop('Churn',axis=1)
# print(x)
x
```

```python
y=df.Churn
y
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2)
```

### Decision Tree Classifier

```python
model_dt = DecisionTreeClassifier(criterion='gini',random_state=100,max_depth=6,min_samples_leaf=10)#one criteria is gini index and one is entropy..
```

```python
model_dt.fit(x_train,y_train)
```

```python
y_pred= model_dt.predict(x_test)
y_pred
```

```python
# model_dt.score(y_test,y_pred)
print(accuracy_score(y_test,y_pred))
```

```python
rfc=RandomForestClassifier(n_estimators=100,criterion='gini',random_state=4)
gbc=GradientBoostingClassifier(n_estimators=100,random_state=4)
lrc=LogisticRegression(solver='sag',penalty='l2')
```

```python
models={
    'LRC': lrc,
    'DT' : model_dt,
    'RFC': rfc,
    'GBC': gbc
    
}
```

```python
def train_classifier(model,x_train,y_train,x_test,y_test):
    model.fit(x_train,y_train)
    y_pred1=model.predict(x_test)
    print('accuracy',accuracy_score(y_test,y_pred1))
    print('roc_score',roc_auc_score(y_test,y_pred1))
    print('classification\n',classification_report(y_test,y_pred1))
    print('confusion_matrix',confusion_matrix(y_test,y_pred1))
```

```python
for name,model in models.items():
    print("For",name)
    train_classifier(model,x_train,y_train,x_test,y_test)
    
```

* GBC is the best 
* Because the data is imabalanced so its giving low precision and accuracy for churn=1 so we will use smott analysis 

```python
# smot analysis is a class to perform over-sampling and cleaning using ENN.
sm=SMOTEENN()
X_resampled,y_resampled = sm.fit_resample(x,y)
```

```python
xr_train,xr_test,yr_train,yr_test= train_test_split(X_resampled,y_resampled,test_size=0.2)
```

```python
models_smote={
    'LRC_smote': lrc,
    'DT_smote' : model_dt,
    'RFC_smote': rfc,
    'GBC_smote': gbc
    
}
```

```python
def train_classifier1(model,x_train,y_train,x_test,y_test):
    model.fit(xr_train,yr_train)
    y_pred1=model.predict(x_test)
    print('accuracy',accuracy_score(y_test,y_pred1))
    print('roc_score',roc_auc_score(y_test,y_pred1))
    print('classification\n',classification_report(y_test,y_pred1))
    print('confusion_matrix',confusion_matrix(y_test,y_pred1))
    
```

```python
for name,model in models_smote.items():
    print("For",name)
    train_classifier1(model,xr_train,yr_train,xr_test,yr_test)
    
```

* best is random forest

```python
rfc_smote = rfc=RandomForestClassifier(n_estimators=100,criterion='gini',random_state=4)
```

```python
train_classifier1(rfc_smote,xr_train,yr_train,xr_test,yr_test)
```
